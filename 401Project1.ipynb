```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")

model.train(
    data="VisDrone.yaml",
    epochs=12,
    imgsz=960,
    batch=8,
    device=0,
    workers=2,
    project="runs_colab",
    name="exp_imgsz960",
    save=True,
    plots=True,
    patience=5,
    verbose=True
)
```

Experiment 1 investigates the impact of increasing input image resolution from 640 to 960.

Compared to the baseline model, the results show a clear improvement:
- mAP50-95 increased from 0.158 to 0.199
- mAP50 increased from 0.282 to 0.338
- Precision increased from 0.407 to 0.465

This improvement is expected because higher resolution images provide more detailed spatial information, which is particularly beneficial for detecting small objects. The VisDrone dataset contains many small and densely packed objects such as pedestrians and bicycles, making resolution an important factor.

However, increasing resolution also significantly increases training time and GPU memory usage. In this experiment, the batch size was reduced to accommodate the larger image size.

Overall, the results demonstrate that input resolution is a critical factor in object detection performance, especially for small-object detection tasks.

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")

model.train(
    data="VisDrone.yaml",
    epochs=12,
    imgsz=640,
    batch=16,
    device=0,
    workers=2,
    project="runs_colab",
    name="exp_aug",
    hsv_h=0.03,
    hsv_s=0.9,
    hsv_v=0.5,
    degrees=5.0
)
```

Experiment 2 investigates the effect of data augmentation on model performance.

Compared to the baseline, the results show a decrease in performance:
- mAP50-95 decreased from 0.158 to 0.130
- mAP50 decreased from 0.282 to 0.238
- Precision and Recall also decreased

This indicates that the applied augmentation strategy negatively impacted the model.

One possible reason is that the VisDrone dataset contains many small objects. Strong augmentations such as high saturation changes (hsv_s=0.9) and brightness variation (hsv_v=0.5) may distort important visual features, making it harder for the model to learn.

Additionally, small objects are more sensitive to transformations like rotation, which can further degrade performance.

Overall, this experiment demonstrates that augmentation must be carefully tuned, especially for datasets with small objects.

```python
from ultralytics import YOLO

model = YOLO("/content/runs/detect/runs_colab/exp_aug/weights/best.pt")
metrics = model.val(data="VisDrone.yaml")

print("Experiment 2 Results")
print("mAP50-95:", metrics.box.map)
print("mAP50:", metrics.box.map50)
```

Three experiments were conducted to evaluate different strategies for improving object detection performance on the VisDrone dataset.

The baseline model achieved an mAP50-95 of 0.158.

In Experiment 1, increasing the input image resolution from 640 to 960 resulted in a significant improvement, raising mAP50-95 to 0.199. This confirms that higher resolution is beneficial for detecting small objects, which are prevalent in the VisDrone dataset.

In Experiment 2, additional data augmentation was applied, including strong color transformations and rotation. However, this resulted in a decrease in performance, with mAP50-95 dropping to 0.130. This suggests that aggressive augmentation can distort important features, particularly for small objects.

Overall, the results indicate that resolution is a more effective factor than augmentation for this dataset. Careful tuning of augmentation strategies is necessary, especially when dealing with small-object detection tasks.

```python
from ultralytics import YOLO
import pandas as pd

models = ["yolo11n.pt", "yolov8n.pt", "yolov5nu.pt", "yolov10n.pt"]
rows = []

for m in models:
    name = m.replace(".pt", "")
    print(f"\nTraining {m}")

    model = YOLO(m)
    model.train(
        data="VisDrone.yaml",
        epochs=6,
        imgsz=640,
        batch=16,
        project="runs_colab",
        name=f"compare_{name}",
        device=0
    )

    best = YOLO(f"/content/runs/detect/runs_colab/compare_{name}/weights/best.pt")
    metrics = best.val(data="VisDrone.yaml")

    rows.append({
        "model": name,
        "mAP50-95": metrics.box.map,
        "mAP50": metrics.box.map50
    })

df = pd.DataFrame(rows)
df
```

Four YOLO models (YOLO11n, YOLOv8n, YOLOv5nu, and YOLOv10n) were evaluated on the VisDrone dataset using identical training settings (6 epochs, image size 640).

Among the models, YOLO11n achieved the best performance with an mAP50-95 of 0.118 and mAP50 of 0.214. YOLOv8n performed similarly but slightly lower, while YOLOv10n and YOLOv5nu showed weaker performance.

The results indicate that YOLO11n provides the best balance between accuracy and efficiency for this dataset.
